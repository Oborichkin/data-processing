```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from src.data_processing.io.image import Image
```

```python
model = Image.from_file("../data/model.jpg")
model.plot()
model = model.threashold(200)
model.plot()
```

```python
d1 = model.dilate()
d1.plot()
(d1 - model).plot()
```

```python
e1 = model.erode()
(model - e1).plot()
```

```python
noised = model.noise(0.15, "gauss").noise(0.15, "sp")
noised.normalize()
noised.plot()
```

```python
d2 = noised.dilate(kernel_size=4)
d2.normalize()
d2.plot()
```

```python
d2_1 = d2.threashold(d2.mode[0])
d2_1.plot()
```

```python
d2_2 = d2_1.dilate()
d2_2.plot()
(d2_2 - d2_1).plot()
```

```python
e2 = noised.erode(kernel_size=10)
e2.plot()
```

```python
e2_1 = e2.threashold(e2.mode[0]-1)
e2_1.plot()
```

```python
e2_2 = e2_1.dilate(kernel_size=10)
e2_2.plot()
```

```python
e2_3 = e2_2.dilate(kernel_size=5)
e2_3.plot()
(e2_3 - e2_2).plot()
```
