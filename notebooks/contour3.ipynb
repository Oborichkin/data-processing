```python
model = Image.from_file("../data/model.jpg")
model.histogram(64)
noised = model.noise(0.15, "gauss").noise(0.15, "sp")
noised.normalize()
noised.plot()
noised.histogram(64)
```

```python
d2 = noised.dilate(kernel_size=4)
d2.normalize()
d2.plot()
```

```python
d2_1 = median_filter(d2, 5)
d2_1.plot()
```

```python
d2_2 = d2_1.threashold(d2_1.mode[0])
d2_2.plot()
```

```python
d2_3 = d2_2.erode(kernel_size=6)
d2_3.plot()
```

```python
d2_4 = median_filter(d2_3, size=3)
d2_4.plot()
```

```python
d2_5 = d2_4.dilate(kernel_size=9)
d2_5.plot()
```

```python
d2_6 = d2_5.erode(kernel_size=3)
d2_6.plot()
(d2_5 - d2_6).plot()
```
